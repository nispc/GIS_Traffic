## 資料來源
臺北市即時交通資訊網的原始碼(http response)抓取
* 臺北市即時交通資訊網 http://its.taipei.gov.tw/
* HTTP message http://its.taipei.gov.tw/ASPX/trafficInfo/trafficInfo.aspx?Lang=CHT&Action=3）

```python
with open('./Taipei_Roads_Data.txt', 'r') as f:
    roads_data = f.readline()
    
roads_data = roads_data.split('&')
roads_data = [i.split(',_') for i in roads_data]
roads_data = roads_data[:-1]
roads_data_df = pd.DataFrame(roads_data, columns=['name', 'speed', 'geometry', 'code', 'start_x', 'start_y', 'end_x', 'end_y', 'dis', 'none'])
roads_data_df = roads_data_df.groupby(by='code', as_index=False).first()
roads_data_df = roads_data_df[['name', 'code', 'geometry', 'start_x', 'start_y', 'end_x', 'end_y']]
roads_data_df['geometry'] = roads_data_df['geometry'].apply(lambda t: [
        (float(j[0]), float(j[1])) for j in (i.split(',') for i in t.split('|')) if t
    ])

roads_data_df = roads_data_df.groupby(by='code', as_index=False).first()
```

```python
roads_data_df['start_x'] = roads_data_df['start_x'].apply(lambda t: float(t) if t else None)
roads_data_df['start_y'] = roads_data_df['start_y'].apply(lambda t: float(t) if t else None)
roads_data_df['end_x'] = roads_data_df['end_x'].apply(lambda t: float(t) if t else None)
roads_data_df['end_y'] = roads_data_df['end_y'].apply(lambda t: float(t) if t else None)
```

## 資料整理

```python
roads_data_df['start_x'] = roads_data_df['start_x'].astype(float)
roads_data_df['start_y'] = roads_data_df['start_y'].astype(float)
roads_data_df['end_x'] = roads_data_df['end_x'].astype(float)
roads_data_df['end_y'] = roads_data_df['end_y'].astype(float)

start_points = [(i, Point(x, y)) for (i, x, y) in zip(roads_data_df.index, roads_data_df['start_x'], roads_data_df['start_y'])]
end_points = [(i, Point(x, y)) for (i, x, y) in zip(roads_data_df.index, roads_data_df['end_x'], roads_data_df['end_y'])]

roads_data_df_start = pd.DataFrame(start_points, columns=['id', 'start']).set_index('id')
roads_data_df_end = pd.DataFrame(end_points, columns=['id', 'end']).set_index('id')

roads_data_df = pd.concat([roads_data_df, roads_data_df_start, roads_data_df_end], axis=1)
```
